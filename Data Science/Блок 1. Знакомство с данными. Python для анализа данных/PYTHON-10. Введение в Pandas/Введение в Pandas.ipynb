```python
#Задание 2.4 (External resource)
import numpy as np
import pandas as pd

names=['chlorhexidine', 'cyntomycin', 'afobazol']
counts=[15, 18, 7]

def create_medications(names, counts):
    medications = pd.Series(index=names, data=counts)
    return medications

def get_percent(medications, name):
    return(medications.loc[name]/sum(medications) * 100)

print(create_medications(names, counts))
#print(get_percent(medications, "chlorhexidine")) #37.5
```

```python
#Задание 3.5 (External resource)
import numpy as np
import pandas as pd

income = [478, 512, 196]
expenses = [156, 130, 270]
years = [2018, 2019, 2020]

def create_companyDF(income, expenses, years):
    df = pd.DataFrame(index=years, data={'Income':income, 'Expenses':expenses})
    return df

#ПРАВИЛЬНЫЙ ОТВЕТ:
def create_companyDF(income, expenses, years):
    df = pd.DataFrame({
        'Income': income,
        'Expenses': expenses
        },
        index = years
    )
    return df

def get_profit(df, year):
    if year in df.index:
        profit = df.loc[year, 'Income'] - df.loc[year, 'Expenses']
    else:
        profit=None
    return profit

print(create_companyDF(income, expenses, years))
print(get_profit(year = 2018, df = create_companyDF([612, 516, 329, 158], [136,163,250,361], [2017,2018,2019,2020])))
```
