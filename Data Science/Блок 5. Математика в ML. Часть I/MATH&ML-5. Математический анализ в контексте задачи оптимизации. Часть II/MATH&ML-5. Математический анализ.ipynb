```python
from sympy import symbols, cos, diff

a, b, c = symbols('a b c', real=True)
f = 5*a*b - a*cos(c) + a**8 + c**2*b

print(diff(f, a))
```

```python
a, b, c = symbols('a b c', real=True)
f = (5-a-2*b)**2

print(diff(f, a))
```

```python
x, y, u = symbols('x y u', real=True)
f = x**2+3*x*y+(4*y)**2

print(diff(f, x))
```
